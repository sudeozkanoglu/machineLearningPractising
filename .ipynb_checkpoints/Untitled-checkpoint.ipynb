```python
# Read in the CSV into a Pandas DataFrame

import pandas as pd

df = pd.read_csv('housing.csv')
df = df.dropna() #Missing values - It's deleting missing values rows (NaN)

df
```

```python
# Seperate into Train and Test DataFrames

# Bu kodun amacÄ± - Veri setini karÄ±ÅŸtÄ±rdÄ±ktan sonra %90 eÄŸitim %10 test olarak ikiye bÃ¶ler.

# df.sample() - veri setindeki verilerin karÄ±ÅŸtÄ±rÄ±lmasÄ±nÄ± saÄŸlar
# frac = 1 - tÃ¼m veri setini karÄ±ÅŸtÄ±rÄ±r, karÄ±ÅŸÄ±m sÄ±rasÄ±nda hiÃ§bir veri atlanmaz
# random_state = 2 - rastgele karÄ±ÅŸÄ±m iÅŸleminin tekrarlanabilir olmasÄ± iÃ§in bir sabit sayÄ± belirler. Bu sayede aynÄ± kod Ã§alÄ±ÅŸtÄ±rÄ±ldÄ±ÄŸÄ±nda her seferinde aynÄ± karÄ±ÅŸÄ±m elde edilir.
df = df.sample(frac=1, random_state=2) 
train_df = df[:17000]
# reset_index() - verification setindeki indexleri sÄ±fÄ±rlar. Bu, karÄ±ÅŸÄ±m sonrasÄ±nda indexlerin sÄ±rasÄ±z hale gelmesini dÃ¼zeltir.
# drop = True - orijinal index sÃ¼tununu tamamen kaldÄ±rÄ±r. EÄŸer bunu kullanmazsanÄ±z, eski index bir sÃ¼tun olarak saklanÄ±r.
train_df = train_df.reset_index(drop=True)
test_df = df[17000:]
test_df = test_df.reset_index(drop=True)

train_df
```

```python
test_df
```

```python
# Get observed outputs for train and test

# Bu kodun amacÄ±, eÄŸitim ve test veri setindeki hedef deÄŸiÅŸkenin boyutlarÄ±nÄ± kontrol etmemizi saÄŸlar.
# EÄŸitim ve test veri setindeki hedef deÄŸiÅŸkeni ayÄ±rarak;
#  1. y_train: Modelin eÄŸitimi sÄ±rasÄ±nda kullanÄ±lacak gerÃ§ek Ã§Ä±ktÄ±lar.
#  2. y_test: Modelin test sÄ±rasÄ±nda karÅŸÄ±laÅŸacaÄŸÄ± gerÃ§ek Ã§Ä±ktÄ±lar.
# olarak saklanÄ±r. Bu iÅŸlemler, hem eÄŸitim hem de test aÅŸamasÄ±nda modelin doÄŸruluÄŸunu ve baÅŸarÄ±sÄ±nÄ± Ã¶lÃ§mek iÃ§in gereklidir.

# train_df['median_house_value'] - eÄŸitim veri setinden median_house_value sÃ¼tununu seÃ§er. Bu sÃ¼tun, modelin tahmin etmeye Ã§alÄ±ÅŸtÄ±ÄŸÄ± hedef deÄŸiÅŸkendir.
# to_numpy() - Pandas veri Ã§erÃ§evesini (DataFrame) bir NumPy dizisine (numpy array) dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r. NumPy dizileri, makine Ã¶ÄŸrenmesi modellerinde daha yaygÄ±n olarak kullanÄ±lÄ±r.
# SonuÃ§ olarak y_train deÄŸiÅŸkeni, eÄŸitim veri setindeki hedef deÄŸiÅŸkenin NumPy dizisini iÃ§erir.
y_train = train_df['median_house_value'].to_numpy()

# test_df['median_house_value'] - test veri setinden hedef deÄŸiÅŸkeni alÄ±r ve Numpy dizisine dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r.
# y_test deÄŸiÅŸkeni, test veri setindeki hedef deÄŸiÅŸkeni iÃ§erir.
y_test = test_df['median_house_value'].to_numpy()

# .shape - NumPy dizisinin boyutunu dÃ¶ndÃ¼rÃ¼r. Ã–rneÄŸin; eÄŸer y_train 17.000 eleman iÃ§eriyorsa sonucu (17000,) ÅŸeklinde verir. EÄŸer y_test 3433 eleman iÃ§eriyorsa, sonucu (3433,) ÅŸeklinde verir.
y_train.shape, y_test.shape
```

```python
# Get error of baseline model that uses the average median_house_value in train_df

# Bu kod, temel olarak median_house_value ortalamasÄ±nÄ± kullanÄ±r. Her test Ã¶rneÄŸi iÃ§in bu sabit deÄŸeri tahmin eder. mean_absolute_error fonksiyonu, bu sabit tahminlerin gerÃ§ek deÄŸerlerden ne kadar saptÄ±ÄŸÄ±nÄ± Ã¶lÃ§er.
from sklearn.metrics import mean_absolute_error

# train_df['median_house_value'] - eÄŸitim veri setindeki hedef deÄŸiÅŸken sÃ¼tunu seÃ§ilir.
# .mean() - Bu sÃ¼tundaki deÄŸerlerin aritmetik ortalamasÄ±nÄ± hesaplar.
# SonuÃ§ - average_median_house_value, temel modelin kullanacaÄŸÄ± sabit tahmin deÄŸeridir (yani tÃ¼m tahminlerde bu deÄŸer kullanÄ±lacak).
average_median_house_value = train_df['median_house_value'].mean()

# [average_median_house_value] - ortalama deÄŸer, bir listeye dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lÃ¼r
# * len(test_df) - test veri setindeki Ã¶rnek sayÄ±sÄ± kadar aynÄ± tahmini (ortalama deÄŸeri) iÃ§eren bir liste oluÅŸturur.
# SonuÃ§ - baseline_model_test_predictions, test veri setindeki her bir Ã¶rnek iÃ§in aynÄ± tahmini (ortalama deÄŸer) yapar.
baseline_model_test_predictions = [average_median_house_value] * len(test_df)

# mean_absolute_error - gerÃ§ek deÄŸerler (y_test) ile tahmin edilen deÄŸerler (baseline_model_test_predictions) arasÄ±ndaki ortalama mutlak hata (MAE) deÄŸerini hesaplar.
# AmaÃ§ - modelin performansÄ±nÄ± Ã¶lÃ§mek
mean_absolute_error(baseline_model_test_predictions, y_test)
```

```python
# PCA
# Observe correlation between total_rooms, total_bedrooms and households

# train_df[['total_rooms', 'total_bedrooms', 'households']] - eÄŸitim veri setinden yalnÄ±zca bu sÃ¼tunlarÄ± seÃ§er
# .corr() - seÃ§ilen sÃ¼tunlar arasÄ±nda Pearson korelasyon katsayÄ±larÄ±nÄ± hesaplar
# Pearson Korelasyon KatsayÄ±sÄ± - Ä°ki deÄŸiÅŸkenin arasÄ±ndaki doÄŸrusal iliÅŸkinin gÃ¼cÃ¼nÃ¼ ve yÃ¶nÃ¼nÃ¼ ifade eder.
#  DeÄŸer aralÄ±ÄŸÄ± [-1,1] -> +1 mÃ¼kemmel pozitif doÄŸrusal iliÅŸki (biri artarken diÄŸeri de artar)
#                       -> -1 mÃ¼kemmel negatif doÄŸrusal iliÅŸki (biri artarken diÄŸeri azalÄ±r)
#                       -> 0 hiÃ§bir doÄŸrusal iliÅŸki yoktur
# Bu korelasyon analizi, deÄŸiÅŸkenlerin birbirleriyle ne kadar iliÅŸkili olduÄŸunu anlamak iÃ§in yapÄ±lÄ±r. EÄŸer deÄŸiÅŸkenler yÃ¼ksek oranda korelasyona sahipse bu durum ÅŸunlarÄ± gÃ¶sterebilir:
#  DeÄŸiÅŸkenler arasÄ±nda multicollinearity sorunu olabilir
#  PCA gibi boyut indirgeme yÃ¶ntemleri kullanÄ±larak bu deÄŸiÅŸkenler tek bir bileÅŸende temsil edilebilir.
train_df[['total_rooms', 'total_bedrooms', 'households']].corr()
```

```python
# Get X_train_3 which has total_rooms, total_bedrooms and households as inputs

import numpy as np

X_train_3 = train_df[['total_rooms', 'total_bedrooms', 'households']].to_numpy()

X_train_3.shape
```

```python
# Get X_train_3 which has total_rooms, total_bedrooms and households as inputs

X_test_3 = test_df[['total_rooms', 'total_bedrooms', 'households']].to_numpy()

X_test_3.shape
```

```python
# Get error of random forest training on just (X_train_3, y_train) and testing on (X_test_3, y_test)

from sklearn.ensemble import RandomForestRegressor

forest_base = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_3, y_train)
forest_base_test_predictions = forest_base.predict(X_test_3)
mean_absolute_error(y_test, forest_base_test_predictions)
```

```python
# Train PCA (Linear Dimensionality Reduction) with 2-feature output

from sklearn.decomposition import PCA

pca = PCA(n_components= 2)
pca = pca.fit(X_train_3)

pca
```

```python
# Use PCA to transform X_train_3 and X_test_3 into X_train_pca and X_test_pca

# EÄŸitim verisi, PCA ile dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lÃ¼r.
# X_train_3 ve X_test_3, orijinal 3 boyutlu uzaydan, 2 boyutlu bir uzaya projekte edilir.
X_train_pca = pca.transform(X_train_3)
X_test_pca = pca.transform(X_test_3)

X_train_pca.shape, X_test_pca.shape
```

```python
# Get error of Random Forest training on (X_train_pca, y_train) and testing pn (X_test_pca, y_test)

forest_pca = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_pca, y_train)
forest_pca_test_predictions = forest_pca.predict(X_test_pca)

mean_absolute_error(y_test, forest_pca_test_predictions)
```

```python
# Test different preprocessing functions on X_train_3

from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler
import matplotlib.pyplot as plt

# Bu kodun amacÄ±, verileri Ã¶lÃ§eklendirmek (standartlaÅŸtÄ±rmak) ve ardÄ±ndan bir Ã¶zelliÄŸin Ã¶lÃ§eklendirilmiÅŸ deÄŸerinin histogramÄ±nÄ± gÃ¶rselleÅŸtirmektir.
# StandardScaler() - her Ã¶zelliÄŸin ortalamasÄ±nÄ± 0, standart sapmasÄ±nÄ± 1 olacak ÅŸekilde Ã¶lÃ§eklendirir.
# .fit(X_train_3) - eÄŸitim verisi kullanÄ±larak her bir Ã¶zelliÄŸin ortalamasÄ± ve standart sapmasÄ± hesaplanÄ±r.
#   YalnÄ±zca eÄŸitim seti Ã¼zerinde fit iÅŸlemi yapÄ±lÄ±r; test seti bu bilgilere gÃ¶re Ã¶lÃ§eklendirilir (data leakage'Ä± Ã¶nlemek iÃ§in)
scaler = StandardScaler().fit(X_train_3)
X_train_3_scaled = scaler.transform(X_train_3)

# X_train_3_scaled[:, 0] ifadesi, Ã¶lÃ§eklendirilmiÅŸ verinin ilk Ã¶zelliÄŸini alÄ±r. Bu Ã¶zelliÄŸin daÄŸÄ±lÄ±mÄ±nÄ± histogram olarak Ã§izer.
#   Histogram - Ã¶zelliÄŸin Ã¶lÃ§eklendirilmiÅŸ deÄŸerlerinin daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶rselleÅŸtirir. Ortalama 0'a yakÄ±n bir tepe nokta ve 1'e yakÄ±n bir standart sapma beklenir.
plt.hist(X_train_3_scaled[:, 0])
```

```python
# Bu kod, veiryi normalize etmek ve normalize edilmiÅŸ ilk Ã¶zelliÄŸin deÄŸerlerinin histogramÄ±nÄ± Ã§izmeyi amaÃ§lar.
# Normalizer() - Her veri Ã¶rneÄŸinin (satÄ±rÄ±nÄ±n) normunu 1 yapacak ÅŸekilde deÄŸerini Ã¶lÃ§eklendirir.
#   Bu iÅŸlem satÄ±r bazÄ±nda Ã§alÄ±ÅŸÄ±r ve Ã¶zellikler arasÄ±ndaki oranÄ± korur.
#   Bu iÅŸlem genellikle mesafeye dayalÄ± algoritmalarda kullanÄ±lÄ±r. (KNN, K-Means)
normalizer = Normalizer().fit(X_train_3)
X_train_3_normalizer = normalizer.transform(X_train_3)

plt.hist(X_train_3_normalizer[:, 0])
```

```python
# Bu kod, veriyi Min-Max Ã¶lÃ§eklendirme yÃ¶ntemiyle normalize eder ve ardÄ±ndan Ã¶lÃ§eklendirilmiÅŸ ilk Ã¶zelliÄŸin deÄŸerinin histogramÄ±nÄ± Ã§izer.
# MinMaxScaler() - Verilen bir veri setindeki her Ã¶zelliÄŸi, belirli bir aralÄ±ÄŸa (genelde [0,1] Ã¶lÃ§ekler)
#  Bu iÅŸlem, Ã¶zellikle mesafeye dayalÄ± algoritmalarda (KNN, K-Means) veya gradyan tabanlÄ± modellerde kullanÄ±lÄ±r.
minMaxScaler = MinMaxScaler().fit(X_train_3)
X_train_3_minMaxScaler = minMaxScaler.transform(X_train_3)

plt.hist(X_train_3_minMaxScaler[:, 0])
```

```python
# Preprocess X_test_3

X_test_3_scaled = scaler.transform(X_test_3)

X_test_3_scaled.shape
```

```python
# Get error of Random Forest training on (X_train_scaled, y_train) and testing on (X_test_scaled, y_test)

random_forest_scaled = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_3_scaled, y_train)
random_forest_scaled_test_predictions = random_forest_scaled.predict(X_test_3_scaled)

mean_absolute_error(y_test, random_forest_scaled_test_predictions)
```

```python
# Create Pipeline of Standardization -> PCA -> RandomForest(1)

# Bu kod, bir Pipeline(iÅŸlem hattÄ±) oluÅŸturur ve sÄ±rasÄ±yla normalizasyon, PCA ile boyut indirgeme ve Random Forest Regressor modelini iÃ§erir.
# Pipeline - Veri Ã¶n iÅŸleme ve model oluÅŸturma adÄ±mlarÄ±nÄ± sÄ±ralÄ± bir ÅŸekilde birleÅŸtiren yapÄ±
#   Veri akÄ±ÅŸÄ± otomatikleÅŸtirilir ve her adÄ±m sÄ±rayla uygulanÄ±r.
#   Veri iÅŸleme ve modelleme kodlarÄ±nÄ± bir arada tutarak daha temiz ve hatasÄ±z bir yapÄ± saÄŸlar.
# Avantajlar - Basitlik: Veri Ã¶n iÅŸleme ve modelleme adÄ±mlarÄ± bir arada tanÄ±mlanÄ±r.
#            - Hata Azaltma: TÃ¼m adÄ±mlar belirli bir sÄ±rayla ve tutarlÄ± ÅŸekilde uygulanÄ±r.
#            - Tekrar KullanÄ±m: PipeÅŸine, farklÄ± veri setleri veya projelerde tekrar kullanÄ±labilir.

from sklearn.pipeline import Pipeline

scale_pca_pipe_forest = Pipeline(steps=[('Scaler', StandardScaler()),
                                       ('PCA', PCA(n_components=2)),
                                       ('Forest', RandomForestRegressor(n_estimators=50, max_depth=5))])

scale_pca_pipe_forest
```

```python
# Create Pipeline of Standardization -> PCA -> RandomForest(1)

scale_pca_pipe_forest.fit(X_train_3, y_train)
scale_pca_pipe_forest_test_predictions = scale_pca_pipe_forest.predict(X_test_3)

mean_absolute_error(y_test, scale_pca_pipe_forest_test_predictions)
```

```python
# Create Pipeline of Normalization -> PCA -> RandomForest(1)

normalization_pca_pipe_forest = Pipeline(steps=[('Normalizer', Normalizer()),
                                       ('PCA', PCA(n_components=2)),
                                       ('Forest', RandomForestRegressor(n_estimators=50, max_depth=5))])

normalization_pca_pipe_forest

normalization_pca_pipe_forest.fit(X_train_3, y_train)
normalization_pca_pipe_forest_test_predictions = normalization_pca_pipe_forest.predict(X_test_3)

mean_absolute_error(y_test, normalization_pca_pipe_forest_test_predictions)
```

```python
# Create Pipeline of MinMaxScaler -> PCA -> RandomForest(1)

minMaxScaler_pca_pipe_forest = Pipeline(steps=[('MinMaxScaler', MinMaxScaler()),
                                       ('PCA', PCA(n_components=2)),
                                       ('Forest', RandomForestRegressor(n_estimators=50, max_depth=5))])

minMaxScaler_pca_pipe_forest

minMaxScaler_pca_pipe_forest.fit(X_train_3, y_train)
minMaxScaler_pca_pipe_forest_test_predictions = minMaxScaler_pca_pipe_forest.predict(X_test_3)

mean_absolute_error(y_test, minMaxScaler_pca_pipe_forest_test_predictions)
```
