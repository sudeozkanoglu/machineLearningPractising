```python
# Read in the CSV into a Pandas DataFrame

import pandas as pd

df = pd.read_csv('housing.csv')
df = df.dropna() #Missing values - It's deleting missing values rows (NaN)

df
```

```python
# Seperate into Train and Test DataFrames

# Bu kodun amacı - Veri setini karıştırdıktan sonra %90 eğitim %10 test olarak ikiye böler.

# df.sample() - veri setindeki verilerin karıştırılmasını sağlar
# frac = 1 - tüm veri setini karıştırır, karışım sırasında hiçbir veri atlanmaz
# random_state = 2 - rastgele karışım işleminin tekrarlanabilir olması için bir sabit sayı belirler. Bu sayede aynı kod çalıştırıldığında her seferinde aynı karışım elde edilir.
df = df.sample(frac=1, random_state=2) 
train_df = df[:17000]
# reset_index() - verification setindeki indexleri sıfırlar. Bu, karışım sonrasında indexlerin sırasız hale gelmesini düzeltir.
# drop = True - orijinal index sütununu tamamen kaldırır. Eğer bunu kullanmazsanız, eski index bir sütun olarak saklanır.
train_df = train_df.reset_index(drop=True)
test_df = df[17000:]
test_df = test_df.reset_index(drop=True)

train_df
```

```python
test_df
```

```python
# Get observed outputs for train and test

# Bu kodun amacı, eğitim ve test veri setindeki hedef değişkenin boyutlarını kontrol etmemizi sağlar.
# Eğitim ve test veri setindeki hedef değişkeni ayırarak;
#  1. y_train: Modelin eğitimi sırasında kullanılacak gerçek çıktılar.
#  2. y_test: Modelin test sırasında karşılaşacağı gerçek çıktılar.
# olarak saklanır. Bu işlemler, hem eğitim hem de test aşamasında modelin doğruluğunu ve başarısını ölçmek için gereklidir.

# train_df['median_house_value'] - eğitim veri setinden median_house_value sütununu seçer. Bu sütun, modelin tahmin etmeye çalıştığı hedef değişkendir.
# to_numpy() - Pandas veri çerçevesini (DataFrame) bir NumPy dizisine (numpy array) dönüştürür. NumPy dizileri, makine öğrenmesi modellerinde daha yaygın olarak kullanılır.
# Sonuç olarak y_train değişkeni, eğitim veri setindeki hedef değişkenin NumPy dizisini içerir.
y_train = train_df['median_house_value'].to_numpy()

# test_df['median_house_value'] - test veri setinden hedef değişkeni alır ve Numpy dizisine dönüştürür.
# y_test değişkeni, test veri setindeki hedef değişkeni içerir.
y_test = test_df['median_house_value'].to_numpy()

# .shape - NumPy dizisinin boyutunu döndürür. Örneğin; eğer y_train 17.000 eleman içeriyorsa sonucu (17000,) şeklinde verir. Eğer y_test 3433 eleman içeriyorsa, sonucu (3433,) şeklinde verir.
y_train.shape, y_test.shape
```

```python
# Get error of baseline model that uses the average median_house_value in train_df

# Bu kod, temel olarak median_house_value ortalamasını kullanır. Her test örneği için bu sabit değeri tahmin eder. mean_absolute_error fonksiyonu, bu sabit tahminlerin gerçek değerlerden ne kadar saptığını ölçer.
from sklearn.metrics import mean_absolute_error

# train_df['median_house_value'] - eğitim veri setindeki hedef değişken sütunu seçilir.
# .mean() - Bu sütundaki değerlerin aritmetik ortalamasını hesaplar.
# Sonuç - average_median_house_value, temel modelin kullanacağı sabit tahmin değeridir (yani tüm tahminlerde bu değer kullanılacak).
average_median_house_value = train_df['median_house_value'].mean()

# [average_median_house_value] - ortalama değer, bir listeye dönüştürülür
# * len(test_df) - test veri setindeki örnek sayısı kadar aynı tahmini (ortalama değeri) içeren bir liste oluşturur.
# Sonuç - baseline_model_test_predictions, test veri setindeki her bir örnek için aynı tahmini (ortalama değer) yapar.
baseline_model_test_predictions = [average_median_house_value] * len(test_df)

# mean_absolute_error - gerçek değerler (y_test) ile tahmin edilen değerler (baseline_model_test_predictions) arasındaki ortalama mutlak hata (MAE) değerini hesaplar.
# Amaç - modelin performansını ölçmek
mean_absolute_error(baseline_model_test_predictions, y_test)
```

```python
# PCA
# Observe correlation between total_rooms, total_bedrooms and households

# train_df[['total_rooms', 'total_bedrooms', 'households']] - eğitim veri setinden yalnızca bu sütunları seçer
# .corr() - seçilen sütunlar arasında Pearson korelasyon katsayılarını hesaplar
# Pearson Korelasyon Katsayısı - İki değişkenin arasındaki doğrusal ilişkinin gücünü ve yönünü ifade eder.
#  Değer aralığı [-1,1] -> +1 mükemmel pozitif doğrusal ilişki (biri artarken diğeri de artar)
#                       -> -1 mükemmel negatif doğrusal ilişki (biri artarken diğeri azalır)
#                       -> 0 hiçbir doğrusal ilişki yoktur
# Bu korelasyon analizi, değişkenlerin birbirleriyle ne kadar ilişkili olduğunu anlamak için yapılır. Eğer değişkenler yüksek oranda korelasyona sahipse bu durum şunları gösterebilir:
#  Değişkenler arasında multicollinearity sorunu olabilir
#  PCA gibi boyut indirgeme yöntemleri kullanılarak bu değişkenler tek bir bileşende temsil edilebilir.
train_df[['total_rooms', 'total_bedrooms', 'households']].corr()
```

```python
# Get X_train_3 which has total_rooms, total_bedrooms and households as inputs

import numpy as np

X_train_3 = train_df[['total_rooms', 'total_bedrooms', 'households']].to_numpy()

X_train_3.shape
```

```python
# Get X_train_3 which has total_rooms, total_bedrooms and households as inputs

X_test_3 = test_df[['total_rooms', 'total_bedrooms', 'households']].to_numpy()

X_test_3.shape
```

```python
# Get error of random forest training on just (X_train_3, y_train) and testing on (X_test_3, y_test)

from sklearn.ensemble import RandomForestRegressor

forest_base = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_3, y_train)
forest_base_test_predictions = forest_base.predict(X_test_3)
mean_absolute_error(y_test, forest_base_test_predictions)
```

```python
# Train PCA (Linear Dimensionality Reduction) with 2-feature output

from sklearn.decomposition import PCA

pca = PCA(n_components= 2)
pca = pca.fit(X_train_3)

pca
```

```python
# Use PCA to transform X_train_3 and X_test_3 into X_train_pca and X_test_pca

# Eğitim verisi, PCA ile dönüştürülür.
# X_train_3 ve X_test_3, orijinal 3 boyutlu uzaydan, 2 boyutlu bir uzaya projekte edilir.
X_train_pca = pca.transform(X_train_3)
X_test_pca = pca.transform(X_test_3)

X_train_pca.shape, X_test_pca.shape
```

```python
# Get error of Random Forest training on (X_train_pca, y_train) and testing pn (X_test_pca, y_test)

forest_pca = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_pca, y_train)
forest_pca_test_predictions = forest_pca.predict(X_test_pca)

mean_absolute_error(y_test, forest_pca_test_predictions)
```

```python
# Preprocessing / Scaling

X_train_3
```

```python
# Test different preprocessing functions on X_train_3

from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler
import matplotlib.pyplot as plt

# Bu kodun amacı, verileri ölçeklendirmek (standartlaştırmak) ve ardından bir özelliğin ölçeklendirilmiş değerinin histogramını görselleştirmektir.
# StandardScaler() - her özelliğin ortalamasını 0, standart sapmasını 1 olacak şekilde ölçeklendirir.
# .fit(X_train_3) - eğitim verisi kullanılarak her bir özelliğin ortalaması ve standart sapması hesaplanır.
#   Yalnızca eğitim seti üzerinde fit işlemi yapılır; test seti bu bilgilere göre ölçeklendirilir (data leakage'ı önlemek için)
scaler = StandardScaler().fit(X_train_3)
X_train_3_scaled = scaler.transform(X_train_3)

# X_train_3_scaled[:, 0] ifadesi, ölçeklendirilmiş verinin ilk özelliğini alır. Bu özelliğin dağılımını histogram olarak çizer.
#   Histogram - özelliğin ölçeklendirilmiş değerlerinin dağılımını görselleştirir. Ortalama 0'a yakın bir tepe nokta ve 1'e yakın bir standart sapma beklenir.
plt.hist(X_train_3_scaled[:, 0])
```

```python
# Bu kod, veiryi normalize etmek ve normalize edilmiş ilk özelliğin değerlerinin histogramını çizmeyi amaçlar.
# Normalizer() - Her veri örneğinin (satırının) normunu 1 yapacak şekilde değerini ölçeklendirir.
#   Bu işlem satır bazında çalışır ve özellikler arasındaki oranı korur.
#   Bu işlem genellikle mesafeye dayalı algoritmalarda kullanılır. (KNN, K-Means)
normalizer = Normalizer().fit(X_train_3)
X_train_3_normalizer = normalizer.transform(X_train_3)

plt.hist(X_train_3_normalizer[:, 0])
```

```python
# Bu kod, veriyi Min-Max ölçeklendirme yöntemiyle normalize eder ve ardından ölçeklendirilmiş ilk özelliğin değerinin histogramını çizer.
# MinMaxScaler() - Verilen bir veri setindeki her özelliği, belirli bir aralığa (genelde [0,1] ölçekler)
#  Bu işlem, özellikle mesafeye dayalı algoritmalarda (KNN, K-Means) veya gradyan tabanlı modellerde kullanılır.
minMaxScaler = MinMaxScaler().fit(X_train_3)
X_train_3_minMaxScaler = minMaxScaler.transform(X_train_3)

plt.hist(X_train_3_minMaxScaler[:, 0])
```

```python
# Preprocess X_test_3

X_test_3_scaled = scaler.transform(X_test_3)

X_test_3_scaled.shape
```

```python
# Get error of Random Forest training on (X_train_scaled, y_train) and testing on (X_test_scaled, y_test)

random_forest_scaled = RandomForestRegressor(n_estimators=50, max_depth=5).fit(X_train_3_scaled, y_train)
random_forest_scaled_test_predictions = random_forest_scaled.predict(X_test_3_scaled)

mean_absolute_error(y_test, random_forest_scaled_test_predictions)
```

```python
# Create Pipeline of Standardization -> PCA -> RandomForest(1)

# Bu kod, bir Pipeline(işlem hattı) oluşturur ve sırasıyla normalizasyon, PCA ile boyut indirgeme ve Random Forest Regressor modelini içerir.
# Pipeline - Veri ön işleme ve model oluşturma adımlarını sıralı bir şekilde birleştiren yapı
#   Veri akışı otomatikleştirilir ve her adım sırayla uygulanır.
#   Veri işleme ve modelleme kodlarını bir arada tutarak daha temiz ve hatasız bir yapı sağlar.
# Avantajlar - Basitlik: Veri ön işleme ve modelleme adımları bir arada tanımlanır.
#            - Hata Azaltma: Tüm adımlar belirli bir sırayla ve tutarlı şekilde uygulanır.
#            - Tekrar Kullanım: Pipeşine, farklı veri setleri veya projelerde tekrar kullanılabilir.

from sklearn.pipeline import Pipeline

scale_pca_pipe_forest = Pipeline(steps=[('Scaler', StandardScaler()),
                                       ('PCA', PCA(n_components=2)),
                                       ('Forest', RandomForestRegressor(n_estimators=50, max_depth=5))])

scale_pca_pipe_forest
```

```python
# Create Pipeline of Standardization -> PCA -> RandomForest(1)

scale_pca_pipe_forest.fit(X_train_3, y_train)
scale_pca_pipe_forest_test_predictions = scale_pca_pipe_forest.predict(X_test_3)

mean_absolute_error(y_test, scale_pca_pipe_forest_test_predictions)
```

```python
# Create Pipeline of Normalization -> PCA -> RandomForest(1)

normalization_pca_pipe_forest = Pipeline(steps=[('Normalizer', Normalizer()),
                                       ('PCA', PCA(n_components=2)),
                                       ('Forest', RandomForestRegressor(n_estimators=50, max_depth=5))])

normalization_pca_pipe_forest

normalization_pca_pipe_forest.fit(X_train_3, y_train)
normalization_pca_pipe_forest_test_predictions = normalization_pca_pipe_forest.predict(X_test_3)

mean_absolute_error(y_test, normalization_pca_pipe_forest_test_predictions)
```

```python
# Create Pipeline of MinMaxScaler -> PCA -> RandomForest(1)

minMaxScaler_pca_pipe_forest = Pipeline(steps=[('MinMaxScaler', MinMaxScaler()),
                                       ('PCA', PCA(n_components=2)),
                                       ('Forest', RandomForestRegressor(n_estimators=50, max_depth=5))])

minMaxScaler_pca_pipe_forest

minMaxScaler_pca_pipe_forest.fit(X_train_3, y_train)
minMaxScaler_pca_pipe_forest_test_predictions = minMaxScaler_pca_pipe_forest.predict(X_test_3)

mean_absolute_error(y_test, minMaxScaler_pca_pipe_forest_test_predictions)
```
